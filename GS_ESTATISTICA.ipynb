# Questão 2

### A) Análise a frequência de incêndios florestais antes e depois do governo.

```python
df_gov_atual = df.query("Data_Registro >= '2021-01-01'")
df_gov_anterior = df.query("Data_Registro >= '2016-01-01' and Data_Registro < '2021-01-01'")
freq_incendios_atual = df_gov_atual['descricao_tipologia'].value_counts().get('Incêndio Florestal', 0)
freq_incendios_ant = df_gov_anterior['descricao_tipologia'].value_counts().get('Incêndio Florestal', 0)
print(f'Frequência de incêndios florestais (atual): {freq_incendios_atual}')
print(f'Frequência de incêndios florestais (anterior): {freq_incendios_ant}')
```

### B) Gere um gráfico mostrando as diferenças.

```python
periodos = ["df_dov_anterior", "df_gov_atual"]
frequencias = [freq_incendios_ant, freq_incendios_atual]
plt.bar(periodos, frequencias, color=["red", "green"])
plt.ylabel("Frequência de Incêndios Florestais")
plt.title("Comparação da Frequência de Incêndios Florestais Antes e Depois do Governo")
plt.show()
```

### c) Classifique como “Fake” ou “Not Fake”

```python
limite_percentual = 0.20
if freq_incendios_ant > 0:
    variacao = (freq_incendios_atual - freq_incendios_ant) / freq_incendios_ant
else:
    variacao = float("inf")

classificacao = "Not Fake" if variacao >= limite_percentual else "Fake"

print(f"Variação percentual: {variacao:.2%}")
print(f"Classificação: {classificacao}")
```

# Questão 4

### A) Gerar um dataset contendo o histórico em SP do número de instituições de ensino danificadas, destruídas e a soma das duas.

```python
df['ano'] = pd.to_datetime(df['Data_Registro']).dt.year
df_sp = df[df['Sigla_UF'] == 'SP']

df_inst_danificadas = df_sp.groupby('ano')[['DM_Inst_Ensino_Danificadas', 'DM_Inst_Ensino_Destruidas']].sum().round(2)
df_inst_danificadas['Total'] = df_inst_danificadas['DM_Inst_Ensino_Danificadas'] + df_inst_danificadas['DM_Inst_Ensino_Destruidas']
df_inst_danificadas
```

### b)Compare estes percentuais nos períodos antes e depois do governo.

```python
inicio_governo = 2021
antes_governo = 2016

df_antes = df_inst_danificadas[df_inst_danificadas.index == antes_governo]
df_depois = df_inst_danificadas[df_inst_danificadas.index == inicio_governo]

variacao_danificadas = (df_depois["DM_Inst_Ensino_Danificadas"].values[0] - df_antes["DM_Inst_Ensino_Danificadas"].values[0]) / df_antes["DM_Inst_Ensino_Danificadas"].values[0] * 100
variacao_destruidas = (df_depois["DM_Inst_Ensino_Destruidas"].values[0] - df_antes["DM_Inst_Ensino_Destruidas"].values[0]) / df_antes["DM_Inst_Ensino_Destruidas"].values[0] * 100

variacao_total = (variacao_danificadas + variacao_destruidas) / 2

print(f"Variação das instituições danificadas (2016-2021): {variacao_danificadas:.2f}%")
print(f"Variação das instituições destruídas (2016-2021): {variacao_destruidas:.2f}%")
print(f"Variação total (2016-2021): {variacao_total:.2f}%")
```

### c) Classifique como “Fake” ou “Not Fake”

```python
limite_danificadas = 5
limite_destruidas = 7
limite_total = 10

classificacao = "Not Fake" if (
    variacao_danificadas >= limite_danificadas and
    variacao_destruidas >= limite_destruidas and
    variacao_total >= limite_total
) else "Fake"

print(f"Classificação final: {classificacao}")
```
